## Load libraries and Configs

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt 
import cv2
```

```python
model = YOLO("./yolov8n-seg.pt")
```

```python
!nvidia-smi
```

```python
model.train(
    data = "./segmentation.yaml",
    epochs = 20, 
    batch = 5,
    imgsz = 640,
    workers = 2
)
```

## Inference

```python
img_path = "./runs/segment/train4/results.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.Figure(figsize=(10, 10))
plt.imshow(img)
plt.axis("off")
plt.show()
```
